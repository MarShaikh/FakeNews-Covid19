```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#loading data
df = pd.read_csv('../HPO/model_with_spatio_temporal.csv')
```

```python
x=df['epochs']
y=df['val_acc']
ax = sns.kdeplot(x, y, shade=True)
ax.set(xlim=(-5, 65))
```

```python
sns.relplot(x="epochs", y="val_acc", col="last_activation", style='dense',
            kind="line", data=df);
```
